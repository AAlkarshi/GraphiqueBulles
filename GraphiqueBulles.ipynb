```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("camembert.csv", header=0, sep=",", skipinitialspace=True)

categories = df['Categories']
votes = df['Votes']

# Crer le graphique  bulles
plt.figure(figsize=(12, 8))
plt.scatter(categories, votes, s=votes*10, alpha=0.5)  # s dfinit la taille des bulles
plt.title('Rpartition des votes par catgorie (Graphique  bulles)')
plt.ylabel('Votes')
plt.xticks(rotation=45)  # Rotation des labels pour une meilleure lisibilit
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("camembert.csv", header=0, sep=",", skipinitialspace=True)

categories = df['Categories']
votes = df['Votes']

# Créer le graphique à bulles
plt.figure(figsize=(12, 8))
plt.scatter(categories, votes, s=votes*10, alpha=0.5)  # s définit la taille des bulles
plt.title('Répartition des votes par catégorie (Graphique à bulles)')
plt.ylabel('Votes')
plt.xticks(rotation=45)  # Rotation des labels pour une meilleure lisibilité
plt.grid(True)
plt.show()
```
